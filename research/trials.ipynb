```python
data = pd.read_csv('artifacts\data_ingestion\car-pred-data\CarPricesPrediction.csv')
```

```python
data.head()
```

```python
# separation the data type columns [ object and numeric ]

cat = []
num = []
for i in data.columns:
    if data[i].dtypes == 'O':
        cat.append(i)
    else:
        num.append(i)
cat
num
data_n=data[num]   # new dataframe just type numeric
data_c=data[cat]  # new dataframe just type object
```

```python
label_encoders = {}
categorical_columns = data_c.columns  # I would recommend using columns names here if you're using pandas. If you're using numpy then stick with range(n) instead

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    data_c[column] = label_encoders[column].fit_transform(data_c[column])
```

```python
frames = [data_c, data_n]
  
dataF = pd.concat(frames,axis=1)
```
